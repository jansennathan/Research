```python
def h2config(h,bl,t,p): #here I am defining the function that calculates the coordinates
    a = format((bl/2)*np.sin(t*np.pi/180.0)*np.cos(p*np.pi/180.0), '.13g');  #    x
    b = format((bl/2)*np.sin(t*np.pi/180.0)*np.sin(p*np.pi/180.0),'.13g'); # y
    c = format((bl/2)*np.cos(t*np.pi/180.0), '.13g');  #    z
    if h == 1: cord = float(a)
    if h == 2: cord = float(b)
    if h == 3: cord = float(c)
    if abs(float(cord)) < 0.000000000001 or cord == 0: cord = format(0, '.13g')
    return float(cord)
```
